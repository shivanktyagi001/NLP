```python
#nlp Phase -1 Text Processing
#text -> cleaning ->  tokenization -> stpwords removal -> steamming/ Lemmetization -> vectoration
```

```python
text = "i am learning python programming and python is fast!!!!>>>>...."
print("original text :",text)
```

```python
#step - 1: convert to lowercase
text = text.lower()
print("LowerCase :",text)
```

```python
#Step-2: Remove Punctuation - !@#$%^&*()_;'.,
#re - regex-regular expresssion
#r row string python method... to adding \ escpe character so that it red the path and all thing in a correct way
import re
text = re.sub(r'[^\w\s]','',text)
print("Without punctuation: ",text)
#remove everthing except letters and spaces
```

```python
#Step -3: Tokenization (split sentence into words)
tokens = text.split();
print("Tokens : ",tokens)
```

```python
#Step-4:Remove Stopwords - is ,are am,the,an,a
stopwords = ["a","an","am","is","i","the","and"]
filtered_token = []
for word in tokens:
  if word not in stopwords:
    filtered_token.append(word)
print("After removing stopword: ",filtered_token);
```

```python
def stem(word):
  if word.endswith("ing"):
    return word[:-3]
  return word
stemmed_word = []
for word in filtered_token:
  stemmed_word.append(stem(word))
print("After stemming: ",stemmed_word)
```

**Nlp Phase -2 word to numbers**

```python
#will covert text  -> count
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sentences =[
    "i am learning python and python",
    "We are happy today",
    "i am sad today"
]
```

```python
vectorizer1 = CountVectorizer()
x = vectorizer1.fit_transform(sentences)
#learn vocabulary
#convert sentences to numbers
```

```python
print("Vocabulary: ",vectorizer1.get_feature_names_out())
#1.all unique words
#2.sorted in alphabetical order
#3.remove single letters like i,a
```

```python
print("vector:")
print(x.toarray())
#Each row is a sentence
#each colume is a word
#each value is a count
```

#**TF**-**IDF**-- **Term** **frequency** -**Inverse** **Document** **frequency**

```python
#why Tf-Idf
#because count is not enough
#Bag of words treats all word equally
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorize = TfidfVectorizer()
X = vectorize.fit_transform(sentences)
#build vocabulary
#calaculate TF
#calculate idf - impact of each word
#multiple tf * idf
#produce matrix
```

```python
print("vocabulary:",vectorize.get_feature_names_out())
```

```python
print("TF_IDF MAtrix:")
print(X.toarray())
```
